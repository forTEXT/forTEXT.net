## 2. CATMA-Profil laden <a class="anchor" id="second-bullet"></a>

Sie können Ihr CATMA-Profil ganz einfach über ein sogenanntes Access Token für GitMA zugänglich machen. Öffnen Sie dazu Ihren CATMA-Account und klicken Sie auf das Profil-Icon oben rechts (vgl. Abb. 1). Es öffnet sich ein Drop-Down-Menü, in dem Sie "Get Access Token" auswählen können.


**Abb. 1: Access Token im CATMA-Profil anfordern**

Es öffnet sich nun ein Fenster, in dem Sie das Access Token benennen können. Sie können den Zugang Ihres Tokens zeitlich beschränken oder unbeschränkten Zugang gewähren (vgl. Abb. 2). Behalten Sie im Hinterkopf, dass mit diesem Token Zugang zu Ihren gesamten CATMA-Daten gewährt wird. Verwalten Sie in CATMA sensible Daten, so achten Sie darauf, das Access-Token nicht weiterzugeben. Sie können immer wieder neue Access Tokens generieren.

**Abb. 2: Access Token benennen und zeitlich beschränken**

Sie können Access Tokens auch im CATMA-GitLab generieren. Loggen Sie sich dazu [hier](https://git.catma.de/) ein und nutzen Sie die **API**-Settings (vgl. Abb. 3).


**Abb. 3: API-Settings im CATMA-GitLab nutzen**

Geben Sie in der nächsten Box nun Ihr Access Token zwischen den einfachen Anführungszeichen ein. Klicken Sie dann oben im Menü auf "Run".

```python
your_access_token = ''
my_catma = Catma(gitlab_access_token=your_access_token)
```

Um eine Liste Ihrer Projekte einzusehen, klicken Sie in die nächste Zeile und führen Sie sie aus, indem Sie oben im Menü auf "Run" gehen.
